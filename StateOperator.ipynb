## PURE STATES & MIXED STATES

### State vector

$ \vert \Phi \rangle = \alpha_1 \; \vert \varphi_1 \rangle + \alpha_2 \; \vert \varphi_2 \rangle  + ... +   \alpha_n \; \vert \varphi_n \rangle = \sum_{n} \alpha_{n} \vert \varphi_{n} \rangle  \quad$ coherent superposition (quantum probability)

$\sum_{n} \vert \alpha_{n} \vert^2 = 1 $

$ \vert \Phi \rangle \in  \mathcal{H}$

### State operator (density matrix)

***General State***

incomplete information about the system, a system can be described by a state operator 

$ \rho = \sum_{i} p_{i} \vert \Phi_{i} \rangle \langle \Phi_{i} \vert \quad $  general state / incoherent mixture (classical probability ?)

$  \rho  \geqslant 0$

$  tr(\rho)   =  \sum_{n} \vert \alpha_{n} \vert^2 = 1$

$  tr(\rho^2) =  \sum_{n} \vert \alpha_{n} \vert^2 = 1 \quad $ (pure state)

$  tr(\rho^2) =  \sum_{n} \vert \alpha_{n} \vert^2 < 1 \quad $ (mixed state)

<br />

***Pure State***

complete information about the system is available, a system can be described by a vector in Hilbert space as well

$ \rho = \mathcal{P_\Phi} = \vert \Phi \rangle \langle \Phi \vert \quad $  (pure state)


$
\vert \Phi \rangle \langle \Phi \vert =
\begin{pmatrix}
\alpha_1 \\
\alpha_2 \\
... \\
\alpha_n \\
\end{pmatrix} \times
$
$
\begin{pmatrix}
\alpha_1^\dagger & \alpha_2^\dagger & ... & \alpha_n^\dagger \\  
\end{pmatrix} =
$
$
\begin{pmatrix}
\vert \alpha_1\vert^2 & \alpha_1 \alpha_2^\dagger & ... & \alpha_1 \alpha_n^\dagger \\
\alpha_2 \alpha_1^\dagger & \vert \alpha_2\vert^2 & ... & \alpha_2 \alpha_n^\dagger \\
... & ... & ... & ... \\
\alpha_n \alpha_1^\dagger & \alpha_n \alpha_2^\dagger & ... & \vert \alpha_n\vert^2 \\
\end{pmatrix} 
$ 

## TENSOR PRODUCT & ENTANGLED STATE

$ \vert \Phi_A \rangle \otimes \vert \Phi_B \rangle \in \mathcal{H_A} \otimes \mathcal{H_B}$

it describes tensor product of particle A and particle B

$ \vert \Phi \rangle = \alpha_0 \; \vert \varphi_0 \rangle + \alpha_1 \; \vert \varphi_1 \rangle $ 

$ \vert \Psi \rangle = \beta_0 \; \vert \psi_0 \rangle + \beta_1 \; \vert \psi_1 \rangle $ 

$ \vert \Phi \rangle \otimes \vert \Psi \rangle = (\alpha_0 \; \vert \varphi_0 \rangle + \alpha_1 \; \vert \varphi_1 \rangle)
  \otimes (\beta_0 \; \vert \psi_0 \rangle + \beta_1 \; \vert \psi_1 \rangle) 
  = \alpha_0 \beta_0  \; \vert \varphi_0 \rangle \otimes  \vert \psi_0 \rangle  
  + \alpha_0 \beta_1  \; \vert \varphi_0 \rangle \otimes  \vert \psi_1 \rangle  
  + \alpha_1 \beta_0  \; \vert \varphi_1 \rangle \otimes  \vert \psi_0 \rangle
  + \alpha_1 \beta_1  \; \vert \varphi_1 \rangle \otimes  \vert \psi_1 \rangle $

The state vectors that can be written as a tensor product form a subset of $ \mathcal{H}_A \otimes \mathcal{H}_B $

A state vector that cannot be written as  a tensor product is an entangled state

   ### QUANTUM REGISTER AS A TENSOR PRODUCT

$ \vert \varphi \rangle = \vert \varphi_0 \rangle  \otimes  \vert \varphi_1 \rangle \; \otimes ... \otimes \;  \vert \varphi_{n-1} \rangle$

***!!!*** ***Qiskit*** uses an ordering in which the nth qubit is on the left side of the tensor product

$ \vert \varphi \rangle = \vert \varphi_{n-1} \rangle  \otimes  \vert \varphi_1 \rangle \; \otimes ... \otimes \;  \vert \varphi_{0} \rangle$

```python
from matplotlib.cbook import MatplotlibDeprecationWarning

from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex

from math import sqrt
```

```python
qr = QuantumRegister(2, "qbit") #default /0> x /0>
qc = QuantumCircuit(qr)

state = Statevector(qc)

print("\n---- STATE VECTOR ----\n")
display(state.draw('latex'))


state.draw('bloch')
state.draw('hinton')

print("\n-----------------------\n")
```

```python
qr = QuantumRegister(2, "qbit")
qc = QuantumCircuit(qr)

init0 = [1, 0]   # |0> state
init1 = [0, 1]   # |1> state

qc.initialize(init0, 0)
qc.initialize(init1, 1)

state = Statevector(qc)

print("\n---- STATE VECTOR ----\n")
display(state.draw('latex'))

state.draw('qsphere')
state.draw('bloch')
state.draw('hinton')

print("\n-----------------------\n")
```

```python
qr = QuantumRegister(2, "qbit")
qc = QuantumCircuit(qr)

i = 0 + 1.j

init0 = [i/sqrt(2), 1/sqrt(2)]
init1 = [i/sqrt(2), -1/sqrt(2)]

qc.initialize(init0, 0)
qc.initialize(init1, 1)

state = Statevector(qc)

print("\n---- STATE VECTOR ----\n")
display(state.draw('latex'))

state.draw('qsphere')
state.draw('bloch')
state.draw('hinton')

print("\n-----------------------\n")

```

```python
qr = QuantumRegister(2, "qbit")
qc = QuantumCircuit(qr)

qc.h(qr) #Hadamard Gate

state = Statevector(qc)

print("\n---- STATE VECTOR ----\n")
display(state.draw('latex'))

state.draw('qsphere')
state.draw('bloch')
state.draw('hinton')

print("\n-----------------------\n")
```
